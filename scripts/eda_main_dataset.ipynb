```python
sql_injection_bodies = df.loc[
    (df["66 - SQL Injection"] == 1) & df["request_body"].notna(),
    ["timestamp", "request_http_request", "request_body","66 - SQL Injection" , "request_http_method"]
]

sql_injection_bodies
```

```python
pd.set_option("display.max_colwidth", None)

for idx, body in sql_injection_bodies["request_body"].items():
    if "UNION" in str(body):
        print(f"Index: {idx}")
        print(body)
        print("-" * 100)
```

```python
code_injection_bodies = df.loc[
    (df["242 - Code Injection"] == 1) & df["request_body"].notna(),
    ["request_http_request", "request_body"]
]

code_injection_bodies
```

```python
code_injection_bodies["request_http_method"].value_counts()
```

```python
pd.set_option("display.max_colwidth", None)

code_injection_terms = ["<script", "<?php", "{{", "}}", "eval(", "alert("]

for idx, body in code_injection_bodies["request_body"].items():
    if any(term.lower() in str(body).lower() for term in code_injection_terms):
        print(f"Index: {idx}")
        print(body)
        print("-" * 100)
```
